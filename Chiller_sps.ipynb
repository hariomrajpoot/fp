# [ODS].[uspChillerAccountLoad]

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerAccountLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



CREATE   PROCEDURE  [ODS].[uspChillerAccountLoad]
AS   
BEGIN
	
				
IF OBJECT_ID(N'tempdb..#ChillerAccount') IS NOT NULL
BEGIN
DROP TABLE #ChillerAccount;
END
			
 

select distinct  
		a.Id Salesforceid,
		a.Status__c  as  "Account Status",
		a.Name "Account Name",
		d.Name "Account Child Of",
		a.Parent_Account__c "Account Parent",
		au.Email as Account_Owner__c,
		a.Store_number__c,
		a.ShippingStreet,
		a.ShippingCity,
		a.ShippingState,
		a.ShippingPostalCode,
		a.ShippingCountry,
		a.Phone,
		a.TD_Linx__c,
		ap.Email as Retail_Audit_Alert_Recipient__c,
		a.Retail_Broker_Responsibility__c,
		a.[Ghost_Account__c],
		a.[ShippingLatitude] ,
		a.[ShippingLongitude] ,

		HashBytes('MD5',CONCAT(a.Id,
		a.Status__c,
		a.Name ,
		d.Name ,
		a.Parent_Account__c,
		au.Email,
		a.Store_number__c,
		a.ShippingStreet,
		a.ShippingCity,
		a.ShippingState,
		a.ShippingPostalCode,
		a.ShippingCountry,
		a.Phone,
		a.TD_Linx__c,
		ap.Email,
		a.Retail_Broker_Responsibility__c,
		a.Ghost_Account__c,
		a.ShippingLatitude ,
		a.ShippingLongitude)) as Chillerhashkey,

		HashBytes('MD5',CONCAT(a.Id,
		a.Status__c,
		a.Name ,
		d.Name ,
		a.Parent_Account__c,
		au.Email,
		a.Store_number__c,
		a.ShippingStreet,
		a.ShippingCity,
		a.ShippingState,
		a.ShippingPostalCode,
		a.ShippingCountry,
		a.Phone,
		a.TD_Linx__c,
		ap.Email,
		a.Retail_Broker_Responsibility__c,
		a.Ghost_Account__c,
		a.ShippingLatitude ,
		a.ShippingLongitude)) as hashkey


into #ChillerAccount from [Landing].[Chiller_Account] a
left join [Landing].[Chiller_Account] d
on a.ParentId=d.Id
left join [Landing].[Chiller_User] au
on a.Account_Owner__c=au.Name
left join [Landing].[Chiller_User] ap
on ap.Id=a.Retail_Audit_Alert_Recipient__c


where a.Store_number__c is not null 

--and a.Parent_Account__c=d.Parent_Account__c


--select * from [Landing].[Chiller_User] where id='005d0000001pDYQAA2'
--select distinct   Retail_Audit_Alert_Recipient__c from [Landing].[Chiller_Account]


--select top 100 * from #ChillerAccount
MERGE [ODS].[Chiller_Account] AS Target
USING #ChillerAccount	AS Source
ON Source.Salesforceid = Target.Salesforceid



WHEN MATCHED 
AND Source.hashkey<>Target.hash_key THEN
UPDATE SET

	Target.[Salesforceid]=Source.[Salesforceid],
	Target.[Account Status]=Source.[Account Status],
	Target.[Account Name]=Source.[Account Name] ,
	Target.[Account Child Of]=Source.[Account Child Of],
	Target.[Account Parent]=Source.[Account Parent] ,
	Target.[Account Owner]=Source.Account_Owner__c,
	Target.[Store number]=Source.Store_number__c ,
	Target.[ShippingStreet]=Source.[ShippingStreet],
	Target.[ShippingCity]=Source.[ShippingCity],
	Target.[ShippingState]=Source.[ShippingState] ,
	Target.[ShippingPostalCode]=Source.ShippingPostalCode,
	Target.[ShippingCountry]=Source.ShippingCountry,
	Target.[Phone]=Source.Phone ,
	Target.[TD Linx]=Source.TD_Linx__c  ,
	Target.[Retail Audit Alert Recipient]=Source.Retail_Audit_Alert_Recipient__c,
	Target.[Retail Broker Responsibility]=Source.Retail_Broker_Responsibility__c,
	Target.[CoolRExportStatus] = 'U',
	Target.[Update_DateTime] =getdate(),
	Target.[hash_key]=Source.hashkey,
	Target.[Chiller_Hash_Key]=Source.Chillerhashkey,
	Target.[Ghost_Account__c]=Source.Ghost_Account__c,
	Target.[ShippingLatitude]=Source.[ShippingLatitude],
	Target.[ShippingLongitude]=Source.[ShippingLongitude]

WHEN NOT MATCHED BY Target THEN
INSERT ([Salesforceid],
[Account Status],
[Account Name] ,
[Account Child Of] ,
[Account Parent] ,
[Account Owner],
[Store number] ,
[ShippingStreet],
[ShippingCity] ,
[ShippingState] ,
[ShippingPostalCode] ,
[ShippingCountry] ,
[Phone] ,
[TD Linx]  ,
[Retail Audit Alert Recipient],
[Retail Broker Responsibility],
[CoolRExportStatus],
[Inserted_DateTime],
[Chiller_Hash_Key],
[Hash_Key],
[Ghost_Account__c],
[ShippingLatitude],
[ShippingLongitude])
VALUES(

Source.[Salesforceid],

Source.[Account Status],
Source.[Account Name] ,
Source.[Account Child Of] ,
Source.[Account Parent] ,
Source.Account_Owner__c,
Source.Store_number__c ,
Source.[ShippingStreet],
Source.[ShippingCity] ,
Source.[ShippingState] ,
Source.[ShippingPostalCode] ,
Source.[ShippingCountry] ,
Source.[Phone] ,
Source.TD_Linx__c  ,
Source.Retail_Audit_Alert_Recipient__c,
Source.Retail_Broker_Responsibility__c,
'I',
getdate(),
Source.Chillerhashkey,
Source.hashkey,
Source.Ghost_Account__c,
Source.[ShippingLatitude],
Source.[ShippingLongitude]
);


update [ODS].[Chiller_Account]
	set [coolRCode]  =  'F0'+ CAST([Account_Key] as varchar)
	where [coolRCode] is null 
	and Delete_Flag is null 
	and ShippingCountry='USA'

END
GO
```

# [ODS].[uspChillerAssetsLoad]

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerAssetsLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



CREATE   PROCEDURE  [ODS].[uspChillerAssetsLoad]
AS   
BEGIN

				
IF OBJECT_ID(N'tempdb..#ChillerAssets') IS NOT NULL
BEGIN
DROP TABLE #ChillerAssets;
END
			
 

select distinct 
a.Id Salesforceid,
ac.Salesforceid Account_Id,
m.Name AssetType,
a.Status__c,
ac.[Account_Key],
a.Serial_Number__c  Serial_Number,

HashBytes('MD5',CONCAT(a.Id,ac.Salesforceid,m.Name,a.Serial_Number__c,ac.[Account_Key],a.Status__c

)) as Chillerhashkey,

HashBytes('MD5',CONCAT(a.Id,ac.Salesforceid,m.Name,a.Serial_Number__c,ac.[Account_Key],a.Status__c

)) as hashkey


into #ChillerAssets

from [Landing].[Chiller_Assets] a
left join Landing.Chiller_Model m
on a.Model__c=m.Id
left join [ODS].[Chiller_Account] ac
on a.location__c=ac.Salesforceid
where ac.[Store number] is not null





MERGE ODS.[Chiller_Assets] AS Target
USING #ChillerAssets	AS Source
ON Source.Salesforceid = Target.Salesforceid



WHEN MATCHED 
AND Source.hashkey<>Target.hash_key THEN
UPDATE SET

	Target.[Salesforceid]=Source.[Salesforceid],
	Target.[Account_Id] =Source.[Account_Id],
	Target.[StoreCode]=Source.[Account_Key],
	Target.[AssetStatus]=Source.[Status__c],
	Target.[AssetType] = Source.[AssetType],
	Target.[Serial_Number] = Source.[Serial_Number],
	Target.[Update_DateTime] =getdate(),
	Target.[hash_key]=Source.hashkey,
	Target.[Chiller_Hash_Key]=Source.Chillerhashkey

WHEN NOT MATCHED BY Target THEN
INSERT (
[Salesforceid],
[Account_Id],
[StoreCode],
[AssetStatus],
[AssetType] ,
[Serial_Number],
[CoolRExportStatus],
[Inserted_DateTime],
[Chiller_Hash_Key],
[Hash_Key])
VALUES(

Source.[Salesforceid],
Source.[Account_Id],
Source.[Account_Key],
Source.[Status__c],
Source.[AssetType],
Source.[Serial_Number],
'I',
getdate(),
Source.Chillerhashkey,
Source.hashkey
);


END
GO
```

# [ODS].[uspChillerAssetTypeLoad]

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerAssetTypeLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



CREATE   PROCEDURE  [ODS].[uspChillerAssetTypeLoad]
AS   
BEGIN

IF OBJECT_ID(N'tempdb..#assetType') IS NOT NULL
BEGIN
DROP TABLE #assetType;
END


select distinct m.Id
,m.Name,
a.Manufacturer__c as MName, 
m.Manufacturer__c 
as MID into #assetType from 
Landing.Chiller_Model m
left join 
[Landing].[Chiller_Assets] a
on a.Model__c=m.Id



MERGE [ODS].[Chiller_AssetType] AS Target
USING #assetType	AS Source
ON
Target.[Name]=Source.[Name]
--Target.[Id]=Source.[Id]
--select * from #assetType

-- For Updates
WHEN MATCHED 
--AND Source.[Name]<>Target.[Name] 
AND

(Source.[MName] <> Target.[MName] OR Source.[MID] <> Target.[MID] OR  ISNULL(Target.[SalesforceID],'AAA') <> Source.[Id])

THEN
UPDATE SET
Target.[SalesforceID] = Source.[Id],
Target.[MName] = Source.[MName],
Target.[MID]= Source.[MID],
Target.[CoolRExportStatus]='U',
Target.Updated_Datetime = getdate()

-- For Insert
WHEN NOT MATCHED BY Target THEN

INSERT(
	   [Id]
	   ,[Name]
	   ,[MName]
	   ,[MID]
      ,[Inserted_Datetime]
	  ,[SalesforceID]
	)
	VALUES(
	Source.[Id]
	,Source.[Name]
	,Source.[MName]
	,Source.[MID]
	  ,getdate()
	  ,Source.[Id]
	);


	INSERT into [ODS].[Chiller_AssetType] (Name, NewName,[Inserted_Datetime])
	SELECT distinct  s2.Master_Chiller_Model ,s2.Master_Chiller_Model,getdate()
	FROM [ODS].[Chiller_AssetType] s1
	right JOIN [ODS].[Chiller_AssetType_Mapping] s2 ON s1.Name = s2.Master_Chiller_Model 
	where s1.Name is null


	update [ODS].[Chiller_AssetType]
	set Id  =  'AT0000'+ CAST(ChillerAssetType_Key as varchar)
	where Id is null


	UPDATE  a 
	SET a.[NewName] = am.[Master_Chiller_Model]
	from [ODS].[Chiller_AssetType] a
	inner join [ODS].[Chiller_AssetType_Mapping] am
	on a.Name =am.C_1_Model_Name ;

END

GO
```

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerEYCLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



 CREATE PROCEDURE [ODS].[uspChillerEYCLoad]
 as
 BEGIN

 
 INSERT INTO [ODS].[Chiller_Customer_POS] (
		
	   [Customer]
      ,[Product]
      ,[Division]
      ,[store_number]
      ,[Address]
      ,[City]
      ,[State]
      ,[Zip]
      ,[Date]
      ,[Sales_Amount]
      ,[Inserted_DateTime]
	  )
	  select 'Albertson Safeway Parent',eyc.[Product],eyc.[Division] ,eyc.[Store #],eyc.[Address],eyc.[City],eyc.[State],eyc.[Zip],REPLACE(eyc.Week_Date,'E',''),
	  CAST(ISNULL(NULLIF(trim(eyc.Value), ''),0) AS DECIMAL(18,2)),
	  getdate() from [Landing].[Chiller_EYC] eyc
	  left join 
	  [ODS].[Chiller_Customer_POS] pos
	  on 
	  pos.Customer = 'Albertson Safeway Parent' and
	  pos.Product=eyc.Product and
	  pos.[Division] =eyc.[Division] and
	  pos.[store_number] =eyc.[Store #] and
	  pos.[City]=eyc.City and
	  pos.[State]=eyc.[State] and
	  pos.Zip= eyc.[Zip] and
	  pos.[Date]=REPLACE(eyc.Week_Date,'E','')
	  where pos.Date is null



END
GO
```

# [ODS].[uspChillerMarketLoad]

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerMarketLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE  [ODS].[uspChillerMarketLoad]
AS   
BEGIN

IF OBJECT_ID(N'tempdb..#market') IS NOT NULL
BEGIN
DROP TABLE #market;
END

select distinct d.id,d.name into #market from [Landing].[Chiller_Account] a
inner join [Landing].[Chiller_Account] d
on a.ParentId=d.Id and d.ParentId is not  null
where a.Store_number__c is not null  
and a.ShippingCountry='USA' 
and a.Status__c in ('Active','Pending')


MERGE [ODS].[Chiller_Market] AS Target
USING #market	AS Source
ON Target.[Id]=Source.[Id]


-- For Updates
WHEN MATCHED 
AND Source.[Name]<>Target.[Name] 
THEN
UPDATE SET
Target.[Name] = Source.[Name],
Target.[CoolRExportStatus]='U',
Target.Updated_Datetime = getdate()


-- For Insert
WHEN NOT MATCHED BY Target THEN

INSERT(
	   [Id]
	   ,[Name]
      ,[Inserted_Datetime]
	)
	VALUES(
	Source.[Id]
	,Source.[Name]
	  ,getdate()
	);

END

GO
```

# [ODS].[uspChillerShelfLoad]

```sql
/****** Object:  StoredProcedure [ODS].[uspChillerShelfLoad]    Script Date: 08-02-2026 15:14:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [ODS].[uspChillerShelfLoad]
 as
 BEGIN

 INSERT INTO [ODS].[Chiller_Shelf]
 (
 [Date],
 [OutletCode],
 [Main_Serial_Number],
 [Outlet_Name],
 [Classification],
 [Market_Name],
 [TDLinx],
 [Trial],
 [OwnShareOfShelf],
 [ForeignShareOfShelf],
 [EmptyShareOfShelf],
 [Inserted_DateTime]
 )

 select

	   s.[Date]
      ,s.[Outlet Code]
      ,s.[Main Serial Number]
      ,s.[Outlet Name]
      ,s.[Classification]
	  ,s.[Market Name]
	  ,s.[TDLinx]
	  ,s.[Trial]
      ,s.[Own Share of Shelf]
      ,s.[Foreign Share of Shelf]
      ,s.[Empty Share of Shelf]
	  ,getdate()
 
 from [Landing].[Chiller_Shelf] s 
 left join
 [ODS].[Chiller_Shelf] os
 on os.Date=s.Date and
 os.[OutletCode]=s.[Outlet Code] and
 os.[Main_Serial_Number] =s.[Main Serial Number] and
 os.[Outlet_Name] = s.[Outlet Name]  and
 os.[Classification]=s.[Classification]
 where 
 os.Date is null


END
GO
```
